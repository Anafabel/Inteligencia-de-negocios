```python
def fact(x):
    a=x
    if x==1:
        return(a*x)
    elif x==0:
        return(1)
    elif x<0:
        return("Error, los factoriales negativos no existen")
    else:
        return(a*fact(x-1))
```

```python
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y,x,z)
plt.show()
```

```python
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi')  # string must be enclosed with quotes '  '
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)', 'cos(x)'])      # legend entries as seperate strings in a list
plt.show()
```

```python
fig = plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi) # Se asigna un tamaño para la imagen

# Creamos la gráfica del año 1952
data_1952=data[ data.year == 1952 ]
plt.scatter(x=data_1952['lifeExp'], y=data_1952['gdpPercap'])  # Datos para X y Y

plt.show()
```

```python
fig = plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi) # Se asigna un tamaño para la imagen

# Creamos la gráfica del año 1952
tmp=data[ data.year == 1952 ]
plt.scatter(x=tmp['lifeExp'], y=tmp['gdpPercap'])  # Datos para X y Y

# Títulos y escaladores (main and on axis)
plt.yscale('log')
plt.xlabel("Life Expectancy")
plt.ylabel("GDP per Capita")
plt.title("Year: "+str(1952) )
plt.ylim(0,100000)
plt.xlim(30, 90)

plt.show()
```

```python
fig = plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)

# Creamos la gráfica del año 1952
tmp=data[ data.year == 1952 ]
plt.scatter(x=tmp['lifeExp'], y=tmp['gdpPercap'] , # Datos para X y Y
            s=tmp['pop']/200000,                   # Radio escalado de los circulos (población)
            c=tmp['continent'].cat.codes,          # Secuencia de colores de acuerdo al continente
            cmap="Accent",                         # Paleta de colores
            alpha=0.6,                             # Transparencia de los círculos
            edgecolors="white",                    # Color de borde de los marcadores
            linewidth=2)                           # Ancho del borde de los marcadoes

# Add titles (main and on axis)
plt.yscale('log')
plt.xlabel("Life Expectancy")
plt.ylabel("GDP per Capita")
plt.title("Year: "+str(1952) )
plt.ylim(0,100000)
plt.xlim(30, 90)

plt.show()
```

```python
time=data.year.unique()
```

```python
#Se crearán un número de figuras igual al número de años diferentes en la dataframe
for i in time:
    # Instanciamos una figura
    fig = plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)

    # Change color with c and alpha. I map the color to the X axis value.
    tmp=data[ data.year == i ]
    plt.scatter(tmp['lifeExp'], tmp['gdpPercap'] , s=tmp['pop']/200000 , c=tmp['continent'].cat.codes, cmap="Accent", alpha=0.6, edgecolors="white", linewidth=2)

    # Add titles (main and on axis)
    plt.yscale('log')
    plt.xlabel("Life Expectancy")
    plt.ylabel("GDP per Capita")
    plt.title("Year: "+str(i) )
    plt.ylim(0,100000)
    plt.xlim(30, 90)

    # Save it
    filename='Gapminder_step'+str(i)+'.png'
    plt.savefig(filename, dpi=96)
    plt.gca()
```
